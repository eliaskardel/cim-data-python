# Artihmetic Tutorial

```python
counter = 0
ALARM = 100 # this is an constant
```

## int data type 

```python
counter = counter + 10

print(counter) # this prints to the console 
# type(counter) # this prints the data type 
```

```python
'''

this is a multi line statement

'''
if counter >= ALARM:
    print("ALARM")
else:
    print("NO ALARM")
```

## float data type 

```python
counter = counter + 0.1

print(counter) # this prints to the console 
type(counter) # this prints the data type 
```
